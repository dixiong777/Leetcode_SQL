<p>Write an SQL query to find for each user, the join date and the number of orders they made as a buyer in <strong>2019</strong>.</p>

```python
import sqlalchemy
import pandas as pd
engine = sqlalchemy.create_engine('mysql+mysqlconnector://root:newpass_1234@localhost:3306/sys')
```

```python
Users1158 = pd.read_sql_table("Users1158", engine)
Users1158
```

```python
Orders1158 = pd.read_sql_table("Orders1158", engine)
Orders1158
```

```python
query = '''
SELECT u.user_id buyer_id, 
       u.join_date, 
       (SELECT COUNT(DISTINCT o.order_id) orders_in_2019
       FROM Orders1158 o
       WHERE u.user_id = o.buyer_id 
       AND YEAR(order_date) = 2019) orders_in_2019
FROM Users1158 u;
'''
pd.read_sql_query(query, engine)
```
