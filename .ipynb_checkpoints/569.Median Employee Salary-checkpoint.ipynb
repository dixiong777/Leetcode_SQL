```python
import sqlalchemy
import pandas as pd
engine = sqlalchemy.create_engine('mysql+mysqlconnector://root:newpass_1234@localhost:3306/sys')
```

```python
Employee2 = pd.read_sql_table("Employee2", engine)
Employee2
```

```python
query = '''
SELECT ANY_VALUE(e1.Id) Id, e1.Company, e1.Salary
FROM Employee2 e1
LEFT JOIN Employee2 e2
ON e1.Company = e2.Company
GROUP BY e1.Company, e1.Salary
HAVING SUM(CASE WHEN e1.Salary = e2.Salary THEN 1 ELSE 0 END) >=
ABS(SUM(SIGN(e1.Salary - e2.Salary)))
ORDER BY e1.Company, e1.Salary;
'''
```

```python
pd.read_sql_query(query, engine)
```

## Note:
### 1. GROUP BY [Variable, Variable...] HAVING [Condition] ORDER BY  [Variable, Variable...]
### 2. Values:
#### 1) SUM(CASE WHEN e1.Salary = e2.Salary THEN 1 ELSE 0 END): Number of Duplicates^2 (e.g., 1, 4, 9)
#### 2) ABS(SUM(SIGN(e1.Salary - e2.Salary))): Number of larger values - Number of smaller values (e.g. 0, 1, 2, ...)
#### Relationship: 1) = 1 --> 2) = 0/1 for median; 
####                        1) = 4 --> 2) = 0/2 for median;
